### Stemming
Stemming is the process  of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words know as a lemma. Stemming is important in natural lanuage understanding

```python
#  classification problem
# comment of the product is the positive review or negative review
# Reviews ---> eating,eat,eaten [going,goes,gone]-->go


words = ["playing", "played", "plays", "runner", "running", "easily","runs","runed"]

```

# porterstemmer

```python
# porterstemmer
from nltk.stem import PorterStemmer
Porter_Stemmer=PorterStemmer()
```

```python
for word in words:
    print(word,"----->",Porter_Stemmer.stem(word))

Porter_Stemmer.stem('congratulation')

# It will no give the good answer to perform the stemming good
```

RegexpStemmer class

NLTK has regexpstemmer class with the help of which we caneasily implement regular Expression stemmer alogrithms.it basicall takes a single regular expression and remove any prefix or suffix that matches the expresion .let us seen an example

```python
from nltk.stem import RegexpStemmer
Regexp_Stemmer=RegexpStemmer('ing$|s$|e$|able$|ed$', min=4)
for word in words:
    print(Regexp_Stemmer.stem(word))
```

### Snowball stemmer

```python
# it is good as compare to poterstemmer in accuracy
from nltk.stem import SnowballStemmer
snow_ball_stemmer=SnowballStemmer('english')
```

```python
# stemmer the dataset
for word in words:
    print(word,'--->',snow_ball_stemmer.stem(word))
```

```python
# compare betweem the porterstemmer and the snowball stemmer
# snowball stemmer
snow_ball_stemmer.stem('fairly'),snow_ball_stemmer.stem('sportingly')
```

```python
# porterstemmer
Porter_Stemmer.stem('fairly'),Porter_Stemmer.stem('sportingly')
```
