```python
# Insitalized the corpus
corpus=''' Hello welcome,to Dev srivastava home  ture.
please select me to find the best home job in nlp.
'''
print(corpus)
```

```python
# tokenization
# sentence into paragrapgh
from nltk.tokenize import sent_tokenize
Document=sent_tokenize(corpus)
print(Document)
```

```python
# print the sentence in documents;
for sentence in Document:
    print(sentence)
```

```python
# Tokenization
# Paragrapgh --> words
# sentence  -->words
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
# stentence --> words

for sentence in Document:
    print(word_tokenize(sentence))
    
```

```python
#  apply one more word liberies
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)

# it will seprate the puncuation as a different words
```

```python
# tree bank 
from nltk.tokenize import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
tokenizer.tokenize(corpus)

# do no use the middle full stop as the sprate words but ones choice as the seprate full stop
```
