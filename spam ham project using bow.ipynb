Perform the natural language processing (bag of words) to trained the spam ham classification

## BOW,TFIDF,MACHINE LEARNING ALGORITHM

```python
import pandas as pd
# load the dataset
message=pd.read_csv('spam_classification_5441.csv')
message.head()
```

```python
# Data cleaning and preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
Lemmatizer=WordNetLemmatizer()
```

```python
corpus=[]

for i in range(0,len(message)):
    review=re.sub('[^a-zA-Z]',' ',message['message'][i])
    review=review.lower()
    review=review.split()
    review=[Lemmatizer.lemmatize(word,pos='v') for word in review if word not in set(stopwords.words('english'))]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

### Create Bag of Words

```python
# Dependent feature
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
y=encoder.fit_transform(message['label'])
y
```

```python
X=corpus
X
```

```python
# train to split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
X_train
```

```python
from sklearn.feature_extraction.text import CountVectorizer
#  for binary bow enable binary=True
bow=CountVectorizer(max_features=75,binary=True,ngram_range=(1,2))
# Indepndent feature
X_train=bow.fit_transform(X_train).toarray()
X_test=bow.transform(X_test).toarray()
X_train.shape
```

```python
X_test.shape
```

```python
y.shape
```

```python
# train the model
from sklearn.naive_bayes import MultinomialNB
spam_dect_model=MultinomialNB()
spam_dect_model.fit(X_train,y_train)
```

```python
# make the prediction
y_pred=spam_dect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

Using the TFIDF model

```python
# train to split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(corpus,y,test_size=0.30,random_state=42)
```

```python
#  create the TFIDF MODEL
from sklearn.feature_extraction.text import TfidfVectorizer
TFIDF=TfidfVectorizer(max_features=2500,ngram_range=(1,2))
# encode the dataset spam meassage
X_train=TFIDF.fit_transform(X_train).toarray()
X_test=TFIDF.transform(X_test).toarray()
```

```python
# dependent variable are
y
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detection=MultinomialNB()
spam_detection.fit(X_train,y_train)
```

```python
# predict the model
y_pred=spam_detection.predict(X_test)
```

```python
# evaluate the model
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```
