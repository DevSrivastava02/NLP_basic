Perform the stopwords on the DR APJ Abdul Kalam

```python
corpus='''My dear young friends, I have always believed that dreams 
transform into thoughts and thoughts result in action. To achieve great
 things in life, you must dream, and then work hard to turn your dreams
  into reality. Small aim is a crime; have a great aim.
You should always have courage to think differently, to invent, 
to discover, and to travel the unexplored path. Great people are
 not born great, they become great through their hard work, honesty,
  and determination.
Failure will come, but do not fear it. Learn from your failures and 
move forward with confidence. Every failure teaches us a valuable
 lesson and brings us closer to success. Remember, success is not
  a one-time event; it is a continuous journey.
Knowledge and creativity are the true wealth of any nation. I 
always tell the youth of India — never stop learning, because learning
 gives you creativity, creativity leads to thinking, thinking provides
  knowledge, and knowledge makes you great.

My message to all of you is simple: dream big, work hard, stay humble,
 and never give up. If you are determined and believe in yourself, then 
 nothing is impossible. Your efforts today will shape the future of our nation tomorrow.'''
```

```python
# library
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
# stopwords.words('english')
```

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
# perform the sent_tokenize
from nltk.tokenize import sent_tokenize
documents=sent_tokenize(corpus)
type(documents)
```

```python
# apply stopwords and filter and then apply stemming


for i in range(len(documents)):
    # word_tokenize
    words=nltk.word_tokenize(documents[i])
    words=[stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    documents[i]=' '.join(words) # converting all the list of words into sentences
```

```python
# use the snowball stemming
from nltk.stem import SnowballStemmer
snow_ball_stemmer=SnowballStemmer('english')
from nltk.corpus import stopwords
for i in range(len(documents)):
    words=nltk.word_tokenize(documents[i])
    words=[snow_ball_stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    documents[i]=" ".join(words)
```

```python
# use the lemmatization
from nltk.stem import WordNetLemmatizer
word_lemma=WordNetLemmatizer()
from nltk.corpus import stopwords
```

```python
for i in range(len(documents)):
    words=nltk.word_tokenize(documents[i])
    words=[word_lemma.lemmatize(word.lower(),pos='v') for word in words if word not in set(stopwords.words('english'))]
    documents[i]=' '.join(words)
```
